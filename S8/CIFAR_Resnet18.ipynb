```python
from models import *
from utils import *
from data_loaders import *
from train import *
import validation as test
from graph import *
```

```python
classes = ('plane', 'car', 'bird', 'cat',
           'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
BATCH_SIZE = 128
EPOCHS = 20
device = getDevice()
dataloader_args = getDataLoaderArgs(BATCH_SIZE)
trainloader = getTrainDataLoader(dataloader_args, rotation=7.0, verticalFlip=0.25, horizontalFlip=0.25)
testloader = getTestDataLoader(dataloader_args)
model = ResNet18().to(device)
summary(model, input_size=(3, 32, 32))

```

```python
from torch.optim.lr_scheduler import StepLR, CyclicLR
optimizer = getOptimizer(model, weight_decay=0.0005)
scheduler = CyclicLR(optimizer, base_lr=0.01, max_lr=0.1)
for epoch in range(EPOCHS):
  print('EPOCH : ', epoch)
  train(model, device, trainloader, optimizer)
  scheduler.step()
  test.__test(model, device, testloader)
  
```
